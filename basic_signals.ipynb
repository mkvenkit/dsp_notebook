# Basic Signals

This notebook explores basic signals.


```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

Let us look at some Sine and Cosine signals.

```python
SR = 8000     # sampling rate in Hz
SL = 1        # total length of signal in seconds
NS = SR*SL    # total number of samples
A = 1         # amplitude of the signal 

f1 = 100      # frequency of signal in Hz
f2 = 50

t = np.arange(0, NS)
y = 2*np.sin(2.0*math.pi*f1*t/NS) + 3*np.cos(2.0*math.pi*f2*t/NS)

plt.plot(t[:1000], y[:1000])
```

Now let's look at the FFT of this signal.

```python
np.fft.rfft()
```
